1. For the suburb of Altona, it is postulated that a typical property sells for $800,000.
• Use the data at hand to test this assumption.
• Is the typical property price really 800000 or has it increased?
• Use a significance level of 5%.

```python
import pandas as pd
import math
from scipy import stats

# Load dataset
df = pd.read_csv("/Users/dholakiyan/De/property.csv")

# Filter data for Altona suburb
altona = df[df["Suburb"] == "Altona"]

# Get property prices
prices = altona["Price"].dropna()

# Sample statistics
n = len(prices)
mean_price = prices.mean()
std_dev = prices.std(ddof=1)

# Hypothesized mean
mu = 800000

# t-test statistic
t_stat = (mean_price - mu) / (std_dev / math.sqrt(n))

# One-tailed p-value
p_value = 1 - stats.t.cdf(t_stat, df=n-1)

print("Sample size:", n)
print("Mean price:", mean_price)
print("Standard deviation:", std_dev)
print("t-statistic:", t_stat)
print("p-value:", p_value)
```

Hypothesis Test

H₀: Typical property price = $800,000
H₁: Typical property price > 800,000
α = 0.05
p-value = 0.1537 > 0.05
We fail to reject the null hypothesis.

At the 5% significance level, there is not enough statistical evidence to conclude that the typical property price in Altona is greater than $800,000.

The sample mean is $834,830, but the increase is not statistically significant.

2. For the year 2016, is there any difference in the prices of properties sold in the summer months vs winter months?
• Consider months from October till March as winter months and rest as summer months.
• Use a significance level of 5%.

Hypothesis

H₀ (Null Hypothesis): There is no difference in mean property prices between summer and winter.
μ(summer)=μ(winter)

H₁ (Alternative Hypothesis): There is a difference in mean prices.
μ(summer)=!μ(winter)

Significance level:
α=0.05

```python
import pandas as pd
from scipy.stats import ttest_ind
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv("/Users/dholakiyan/Desktop/property.csv")

# Fix date parsing
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

df = df.dropna(subset=["Date"])

# Filter year
df_2016 = df[df["Date"].dt.year == 2016]

# Extract month
df_2016["Month"] = df_2016["Date"].dt.month

# Winter (Oct–Mar)
winter = df_2016[df_2016["Month"].isin([10,11,12,1,2,3])]["Price"]

# Summer (Apr–Sep)
summer = df_2016[~df_2016["Month"].isin([10,11,12,1,2,3])]["Price"]

# t-test
t_stat, p_value = ttest_ind(summer, winter, equal_var=False)

print("Summer mean:", summer.mean())
print("Winter mean:", winter.mean())
print("p-value:", p_value)
```

Using a two-sample t-test at a 5% significance level, the p-value obtained is 2.28×10⁻⁶, which is less than 0.05. Therefore, we reject the null hypothesis and conclude that there is a significant difference in property prices between summer and winter months in 2016, with winter prices being higher.

3. For the suburb of Abbotsford, what is the probability that out of 10 properties sold, 3 will not have a car parking space?
• Use the column car in the dataset.
• Round off your answer to 3 decimal places.

```python
import pandas as pd
from scipy.stats import binom

# Load dataset
df = pd.read_csv("/Users/dholakiyan/Desktop/property.csv")

# Filter suburb Abbotsford
abbotsford = df[df["Suburb"] == "Abbotsford"]

# Probability that a property has NO car parking
p = (abbotsford["Car"] == 0).mean()

# Binomial probability: exactly 3 out of 10
n = 10
k = 3

probability = binom.pmf(k, n, p)

print("Probability of no parking:", p)
print("Final probability:", round(probability, 3))
```

Based on the dataset, the probability that a property in Abbotsford has no car parking space is 0.260. Using the binomial distribution for 10 properties, the probability that exactly 3 properties will not have a car parking space is approximately 0.26 (26%) when rounded to three decimal places.

6. One-Sample Hypothesis Test (Industry Pricing) A real estate firm claims that the average property price in Richmond is $1,000,000. Using the dataset, test whether the actual average price is significantly different from this claim at a 5% significance level. Clearly state:
• Null and alternative hypotheses
• Test statistic
• p-value
• Final business conclusion

Null Hypothesis (H₀):
μ=1,000,000
The average property price in Richmond is $1,000,000.

Alternative Hypothesis (H₁):
𝜇≠1,000,000
The average property price in Richmond is different from $1,000,000.

Significance level:
α=0.05

```python
import pandas as pd
from scipy import stats

# Load dataset
df = pd.read_csv("/Users/dholakiyan/Desktop/property.csv")

# Filter Richmond properties
richmond = df[df["Suburb"] == "Richmond"]["Price"]

# Hypothesized mean
mu = 1000000

# One-sample t-test
t_stat, p_value = stats.ttest_1samp(richmond, mu)

print("Sample mean:", richmond.mean())
print("Test statistic:", t_stat)
print("p-value:", p_value)
```

Conclusion:
At the 5% significance level, the p-value is less than 0.05, indicating strong evidence that the actual average property price in Richmond is significantly different from $1,000,000.

The dataset suggests that the average property price in Richmond is higher than the claimed value of $1,000,000.

7. Independent Two-Sample T-Test (Feature Impact)
Do properties with car parking sell at a higher average price than properties without car parking, across the entire dataset? Use a 5% significance level and justify:
• Choice of test
• Interpretation of p-value
• Business implications for developers

Choice of Test

An Independent Two-Sample t-test is used because:
We are comparing the mean prices of two independent groups:
1. Properties with car parking
2. Properties without car parking
The variable Price is continuous
The groups are independent observations

Hypothesis:
Null Hypothesis (H₀):

μ(parking)=μ(no parking)
The average price of properties with car parking is equal to the average price of properties without car parking.

Alternative Hypothesis (H₁):
μ(parking)>μ(no parking)
Properties with car parking sell at a higher average price.

Significance level:
α=0.05

```python
import pandas as pd
from scipy.stats import ttest_ind

# Load dataset
df = pd.read_csv("/Users/dholakiyan/Desktop/property.csv")

# Separate groups
with_parking = df[df["Car"] > 0]["Price"]
without_parking = df[df["Car"] == 0]["Price"]

# Independent t-test
t_stat, p_value = ttest_ind(with_parking, without_parking, equal_var=False)

print("Mean price with parking:", with_parking.mean())
print("Mean price without parking:", without_parking.mean())
print("t-statistic:", t_stat)
print("p-value:", p_value)
```

Interpretation of p-value
Significance level:
α=0.05

p-value:
p=0.785

Comparison:
0.785>0.05

Decision:
Fail to reject the null hypothesis (H₀).
This means there is no statistically significant difference in average property prices between properties with car parking and those without car parking in the dataset.
The negative t-statistic also indicates that the mean price without parking is slightly higher, but the difference is not statistically significant.

Business Implications for Developers:
Since the statistical test shows no significant price difference, the presence of car parking does not appear to strongly influence property prices in this dataset.

Implications:
Developers should not assume parking automatically increases property value.
Other factors such as location, number of rooms, property size, and neighborhood amenities may have a greater impact on price.
In some areas, buyers may prioritize affordability or proximity to public transport rather than parking availability.
Developers should analyze multiple property features together when determining pricing strategies.

10. Industry-Style Hypothesis Validation (Policy Decision)
A housing policy group believes that properties with more than 2 bathrooms command a premium price.
Design and execute a statistical test to validate this claim:
• Identify the correct test
• State hypotheses
• Report p-value
• Give a clear recommendation to policymakers

1. Identify the Correct Test

The appropriate statistical test is an Independent Two-Sample t-test.

Reason:
We are comparing the average prices of two independent groups:
1. Properties with more than 2 bathrooms
2. Properties with 2 or fewer bathrooms
The variable Price is continuous.
The groups are independent observations.

Significance level:
α=0.05

2. Hypotheses

Null Hypothesis (H₀):
μ>2(bath)≤μ≤2(bath)
Properties with more than 2 bathrooms do not have a higher average price.

Alternative Hypothesis (H₁):
μ>2(bath)>μ≤2(bath)
Properties with more than 2 bathrooms have a higher average price (premium).

```python
import pandas as pd
from scipy.stats import ttest_ind

df = pd.read_csv("/Users/dholakiyan/Desktop/property.csv")

# Create groups
more_than_two = df[df["Bathroom"] > 2]["Price"]
two_or_less = df[df["Bathroom"] <= 2]["Price"]

# Independent t-test
t_stat, p_value = ttest_ind(more_than_two, two_or_less, equal_var=False)

print("Mean price (>2 bathrooms):", more_than_two.mean())
print("Mean price (≤2 bathrooms):", two_or_less.mean())
print("t-statistic:", t_stat)
print("p-value:", p_value)
```

3. p-value Interpretation:

The p-value = 7.38 × 10⁻¹³⁷ is extremely small and much less than 0.05.
7.38 × 10⁻¹³⁷<0.05

Decision:
Reject the null hypothesis (H₀).

Interpretation:
There is very strong statistical evidence that properties with more than 2 bathrooms sell at higher average prices compared to properties with two or fewer bathrooms.

The large t-statistic (28.8) also indicates a strong difference between the two groups.

4. Recommendation to Policymakers
The results suggest that additional bathrooms significantly increase property value.

Implications for policymakers and developers:
1. Housing developments should consider designing homes with more than two bathrooms to increase property value.
2. Policies encouraging larger or higher-quality housing infrastructure may positively influence property market value.
3. Developers may prioritize multi-bathroom housing projects, especially in high-demand urban areas.
Providing more bathrooms appears to be a strong market feature that buyers value, leading to higher selling prices.

Conclusion:
The independent two-sample t-test shows a statistically significant difference in property prices between homes with more than two bathrooms and those with two or fewer bathrooms.
Since the p-value is far below the significance level of 0.05, we reject the null hypothesis and conclude that properties with more than two bathrooms command a premium price.
    The analysis clearly indicates that bathroom count is an important factor influencing property value, and homes with more than two bathrooms sell at significantly higher prices.
